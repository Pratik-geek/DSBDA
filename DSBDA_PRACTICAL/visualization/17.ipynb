```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
cl = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']
df = pd.read_csv('Cleavland.csv', names=cl)
df
```

```python
df.isnull().sum()
```

```python
#Histogram
plt.hist(df['age'],bins=15)
plt.xlabel('AGE')
plt.ylabel('FREQUENCY')
plt.title('HISTOGRAM')
plt.show()
```

```python
plt.plot(df['sex'],df['age'],'go',alpha=0.5)
plt.xlabel('sex')
plt.ylabel('age')
plt.title('DOT PLOT')
plt.show()
```

```python
gc = df['sex'].value_counts()
gc
plt.bar(gc.index,gc.values)
plt.xlabel('Gender')
plt.ylabel('counts')
plt.title('BAR PLOTS')
plt.show()
```

```python
#LINECHART
plt.plot(df['age'],df['trestbps'],'g')
plt.xlabel('AGE')
plt.ylabel('trestbps')
plt.title('LINE CHART')
plt.show()
```

```python
plt.scatter(df['age'],df['trestbps'])
plt.xlabel('AGE')
plt.ylabel('trestbps')
plt.title('LINE CHART')
plt.show()
```

```python
gc = df['sex'].value_counts()
gc
```

```python
plt.pie(gc,labels=gc.index)
plt.title('PIE CHART')
plt.show()
```

```python
#BOX PLOT
plt.boxplot(df['chol'])
plt.xlabel('cholestrol')
plt.title('Box Plot: Cholesterol')
plt.show()
```
